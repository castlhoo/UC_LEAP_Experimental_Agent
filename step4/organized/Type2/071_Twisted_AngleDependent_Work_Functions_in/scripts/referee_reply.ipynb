```python
E0=-0.005
dE=0.0005

energies = []
conductances = []
Gs = []
folder_name = data_dir + f'/Hartree and corrugation/delta_nu=0.000_corrugation_amplitude=1.000/'

for i in range(20):
    E = E0 + i * dE 
    filename = folder_name + f'energy_range_{E:.5f}_{E + dE - 0.0001:.5f}'
    with open(filename ,'rb') as file:
        data = dill.load(file)

    for energy in data[0]:
        energies.append(energy * 1000)
    for conductance in data[1]:
        Gs.append(conductance)
        conductances.append(-conductance[0][1])
```
